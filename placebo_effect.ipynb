#### What is placebo effect ? ####
Do you think it is possible to heal using sugar? Sounds fun or even a bit ironic, right? Suprisingly, it is not impossible. Actually, this is called placebo effect:
That is a type of a medical treatment which uses sugar pills which are told to be actual medicine, to make patients think that they are consuming real medicine and will get healed.

#### History ####
It is time to dive into the history :)
Ironically the person, who first found out about the effect was not part of any medical institution, he was actually an English physician called John Haygard in 1799. He took metal pointers and called it "Perkins tractors", which ware an effective medical treatment for those times. People didn't know that the "Perkins tractors" were fake and actually believed that they would be cured. He sold them on a really high price, but amazingly it worked. Then he wrote a book with a really interesting title - "On the Imagination as a Cause & as a Cure of Disorders of the Body", where he describes this effect. This historical momet proves the power of the human mind and imagination. As the title says, the imagination can be the cause, but also the cure of the body disorders.

At first placebo effect was used just for mental disorders, because it can be described as a brain manipulation. You think and convince yourself that your health is going to improve due to this "medication". When you start to think in this direction without a doubt we can see the huge impact on the brain to your body. These "pills" were used to reduce pain, anxiety, depression, etc, because these symptoms were caused by the brain. There were many contradictions and disbeliefs towards this effect, but after hundreds of experiments, not only from medical faces, but also psychologists, they came up with the idea that our brain is the most powerful tool. It has the power to stimulate healing. A Harvard article about this topic says that if we succeed in convincing our brain that we are about to heal even though we are consuming fake medication it can be as effective as the tradinional one. Placebo effect is a way to make your body feel better, because of the neurotransmitters - dopamine and endorphins, which increase the feeling of happiness and bring us better moods.

Also if we dig deeper we can see the incredible connection between physical and mental health. Some studies show that physical changes can occur using those effect. Our brain is the most complex organ, which is responsible for thousands operations in our body every minute. If you have already convinced the brain that you are healthy, you are halfly cured and you will definitely feel better. There is nothing stronger than the human will, positive thinking and faith. People's expactations can form the effect of the "pill" to their bodies. So the therapeutical results and the psychological outcome depend mostly on patient's mindset and attitude towards the treatment.

I will try to write another algorithm for calculating effectiveness. We have two main variables to take in consideration - the time interval to heal and the recurrent patients.

Since we don't have an information about the number of recurrencies as the dataset above, in my formula I will assume that we have one recurrencies, so that means the patient get healed for some x time, but then he have to return and heal for another x time, so the time intervals grow.We know that time for the first healing is different from the time for the second one, but since we have the mean time interval, we can use it.

So the formula is:
(1 + percentage of recurrent patients) * mean_time_interval
We know that recurrent patients return back and spend time again in healing , so I have to take them in mind when considering the better treatment.And the higher the number, the worse the treatment, because the percentage of recurrent patients and the time to heal are variables, that when are growing exponential is considered to be a negative sign for the effectiveness of the type of treatment.
